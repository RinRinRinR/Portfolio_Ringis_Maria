### Резюме проекта
Стартап разрабатывает SaaS-платформу, которая с помощью машинного обучения анализирует рыночные данные о зарплатах и определяет справедливую оплату труда для конкретных должностей с учетом:

Географии (регион, город)

Отрасли и специфики компании

Квалификации и опыта сотрудника

Рыночных тенденций

Экономических показателей

Ценностное предложение: Оптимизация расходов на персонал при сохранении конкурентоспособности на рынке труда.

## Этапы пользовательского пути и монетизации.Воронка конверсии:
### «Завлечение» (Бесплатно)

**Первый запрос — бесплатно.** Пользователь регистрируется (собираем контактные данные) и получает полноценный отчёт по своему первому запросу.
**Цель:** Доказать ценность, показать качество данных, создать доверие.


### «Монетизация разового использования» (Pay-Per-Use)

После первого бесплатного запроса пользователь переходит на тариф «По запросу».

**Стоимость:** 100 рублей за каждый последующий отчёт.

**Психологический механизм:** Пользователь, получив ценность, с большей вероятностью заплатит небольшую сумму. Он чувствует контроль — платит только когда нужно.


### «Создание трения и демонстрация выгоды» (Upsell)

**Дизайн интерфейса:** После каждого платного запроса (или даже перед его оплатой) появляется ненавязчивое, но чёткое сообщение:

«За этот запрос вы заплатите 100 ₽. С подпиской за 790 ₽ в месяц вы получаете до 10 запросов. Экономьте до 40%!»

**Кабинет пользователя:** В личном кабинете наглядно показывается:

Сколько запросов сделано в этом месяце.

Сколько уже потрачено денег по pay-per-use.

Сколько можно было бы сэкономить с подпиской.

Счётчик: «До выгодного тарифа вам осталось N запросов».


### «Конверсия в подписчика»

**Эффект якоря:** Бесплатный первый запрос создаёт положительный якорь. 100 рублей за запрос кажется маленькой суммой по сравнению с полученной ценностью.

**Боль упущенной выгоды (FOMO):** Видя, сколько он переплачивает, пользователь начинает чувствовать, что «теряет деньги», не переходя на подписку.

**Снижение когнитивной нагрузки:** При активном использовании постоянно думать «стоит ли делать этот запрос за 100 рублей?» утомительно. Подписка снимает это ограничение, что воспринимается как освобождение.

**Путь наименьшего сопротивления:** После оплаты подписки пользователь начинает использовать сервис активнее (эффект «шведского стола»), что увеличивает его вовлечённость и зависимость от сервиса.

## Разработка продукта: этапы PoC, MVP, MLP  
### PoC (Proof of Concept) — «Доказательство работоспособности» (1-2 недели)  

**Цель:** Убедиться, что ядро идеи технически выполнимо и дает ценность.

Что делаем:

**Сбор данных (1 неделя):**

Парсим открытые источники (HH, Superjob, Zarplata.ru) силами основателей или студентов.  
Фокус на 1-2 ключевые профессии (например, Python-разработчик и SMM-специалист) и 2-3 города (Москва, СПб, Екатеринбург).  
Собираем минимальный набор признаков: должность, город, вилка зарплат, опыт, формат работы.  

**Простая модель (3-4 дня):**

Используем базовые алгоритмы (линейная регрессия, дерево решений).  
Цель модели: по введенным параметрам (должность, город, опыт) выдать предсказанную вилку зарплат (25-й, 50-й, 75-й перцентили).  

**Интерфейс (2-3 дня):**

Простейший телеграм-бот или одностраничный сайт на Streamlit.  
Пользователь вводит 3-4 параметра, получает ответ в формате: «В Москве Python-разработчик с опытом 1-3 года может получать 90 000 – 130 000 руб.».  
Результат: Рабочий прототип, который дает грубую, но осмысленную оценку. Основатели убеждаются, что «магия работает».  

### MVP (Minimum Viable Product) — «Минимально жизнеспособный продукт» (3-4 недели)

**Цель:** Получить первый фидбэк от реальных пользователей и проверить гипотезу о спросе.

Что делаем:

**Команда:** Нанятые знакомые из колледжа/вуза (2 разработчика: бэкенд+фронтенд).

**Разработка (1 месяц):**

**Бэкенд:**
 
Автоматизированный пайплайн сбора данных (скрипты парсинга, которые запускаются раз в неделю).  
Модель ML с более широким охватом профессий (10-15) и городов (10-15).  
Добавляем 1-2 фактора: тип компании (стартап, корпорация) или формат (офис/удаленка).  
Простая система пользователей (регистрация по email).  

**Фронтенд (веб-приложение):**

Чистый, минималистичный интерфейс.  
Форма ввода параметров (выпадающие списки, слайдеры).  
Страница с результатом (вилка зарплат, краткая аналитика).  
Личный кабинет: история запросов, счетчик (1 бесплатный, далее платно).  
Экспериментальная страница с тарифами (pay-per-use за 100 руб., подписка за 790 руб./мес), но платежка не интегрируется (заглушка «Скоро!»).  

**Тестирование в колледже (1 неделя):**

Даем доступ 30-50 студентам и преподавателям.  
Собираем фидбэк через Google Forms: понятен ли интерфейс, полезен ли результат, какие параметры не хватает, сколько бы они заплатили.  

**Фиксация результатов:**

Какие профессии/фильтры самые популярные?  
Насколько точной кажется оценка?  
Готовы ли люди платить? Что их смущает?  

**Результат:** Рабочее веб-приложение, первые 50 пользователей, список доработок для MLP, подтверждение/опровержение гипотез.

##  MLP (Minimum Lovable Product) — «Минимально любимый продукт» (2-3 месяца)

**Цель:** Создать продукт, который не просто работает, а нравится первым пользователям и готов к платному запуску.

**Что делаем (после MVP и фидбэка):**

**Команда:** Нанять 1-2 опытных разработчиков (или отдать на аутсорс) для качественной переработки кода.

**Ключевые доработки:**

**Данные и модель:**

Расширить охват (50+ профессий, 30+ городов).

Улучшить модель (ансамбли, feature engineering: учет инфляции, сезонности).

Добавить «индекс востребованности» (соотношение вакансий к кандидатам).

**Интерфейс и UX:**

Проработанный дизайн (Figma → готовый фронтенд).

Интерактивная визуализация: график распределения зарплат, сравнение по городам.

Возможность сохранять и сравнивать несколько сценариев.

**Ядро монетизации:**

Интеграция платежной системы (ЮKassa, CloudPayments).

Рабочие тарифы: Первый запрос бесплатно → Pay-per-use (100 руб.) → Подписка (790 руб./мес, 10 запросов).

Система уведомлений и напоминаний о выгоде подписки.

**Безопасность и стабильность:**

Защита от ботов и накруток.

Резервное копирование данных.



## Пилотный маркетинг («Правило 20/80»):

- Создаем посты на VC.ru, Habr (не как рекламу, а как кейс: «Как мы сделали сервис оценки зарплат»).

- Запускаем таргетированную рекламу в Яндекс.Директ / ВК на малые аудитории:

«HR-менеджеры в Москве, компании 10-100 человек».

«Python-разработчики, ищущие работу».

Бюджет: 10-20 тыс. руб. на тест.

Цель: Получить первые 100 платящих пользователей и понять CAC (стоимость привлечения клиента).

**Сбор метрик:**

Конверсия из посетителя в бесплатного пользователя.

Конверсия из бесплатного в платного (pay-per-use).

Конверсия из pay-per-use в подписку.

Retention (возвращаемость) пользователей.

**Результат:** Рабочий, стабильный и «любимый» продукт, готовый к масштабированию. Первые реальные деньги, понимание unit-экономики, дорожная карта для развития.



## Поэтапная стратегия выхода на рынок: от МСБ к корпорациям и госсектору

### Этап 1: Фундамент и МСБ (0-2 года)*

**Продукт:** Тот самый гибридный сервис (фримиум + pay-per-use + подписка).

**ЦА:** Соискатели, фрилансеры, малый и средний бизнес (до 100 сотрудников).

**Задача:** Накопить данные, отточить алгоритмы, собрать первые 1000-5000 платных пользователей, создать базу кейсов и отзывов.

**Финансы:** Живём на выручку от подписок МСБ и разовых запросов.


### Этап 2: Выход на средний и крупный бизнес (2-4 года)

**Продукт:** B2B-тарифы.

**«Professional»:** Для отдельных HR-ов в больших компаниях. Расширенные фильтры, больше запросов, экспорт в Excel/PPT.

**«Team» / «Corporate»:** Мульти-аккаунты, API для интеграции, кастомные бенчмарки (сравнение с данными по их отрасли), персональный менеджер.

**Каналы продаж:** Прямые продажи (отдел B2B-продаж), партнёрства с HR-агентствами, консалтинговыми фирмами, участие в отраслевых конференциях (HR-Tech, SHRM).

**Задача:** Заключить первые пилотные проекты с известными брендами (не обязательно гигантами, но с узнаваемыми именами). Использовать их логотипы в разделе «Наши клиенты».

    
### Этап 3: Госсектор и монополии (4+ лет)

**Продукт:** Кастомизированное решение.

Учёт гос. категорий и разрядов (Единая тарифная сетка и т.д.).

Анализ зарплат в бюджетной сфере (образование, медицина) и госкомпаниях.

Полная локализация данных и хранения в РФ.

Участие в тендерах (как поставщик аналитического ПО).

**Каналы продаж:** Лоббирование через отраслевые ассоциации, работа с «дочками» госкорпораций, прямые контракты через госзакупки.

**Задача:** Стать стандартом де-факто для анализа рынка труда на государственном уровне.

## финальная цель: Создание собственной биржи/платформы труда


Станем источником правды о зарплатах. К нам будут приходить все, чтобы узнать свою стоимость. Создадим сообщество (соискатели + компании).

Добавим профили. Соискатели, которые оценили свою стоимость, могут захотеть создать «расширенный профиль» с резюме, портфолио, верификацией навыков (наш ИИ может и это анализировать).

Запустим «умный» поиск работы/кандидатов. На основе ваших же данных о рыночной стоимости. Вы сможете предлагать: «Компании, готовой платить 150к+, покажите этого кандидата в первую очередь».

**Монетизация платформы:**

**Для соискателей:** Премиум-профиль, поднятие резюме в поиске, расширенная аналитика.

**Для компаний:** Публикация вакансий, доступ к базе кандидатов, автоматический подбор (AI Matching), аналитика откликов.

## 2. ПРОГНОЗ РОСТА ПОЛЬЗОВАТЕЛЕЙ (12 МЕСЯЦЕВ)
**Методология расчёта:**

**Приток пользователей:** Будет расти экспоненциально с улучшением продукта и маркетинга

**Конверсии:** Основаны на средних показателях SaaS (5-7% в платящие) с учётом специфики продукта

**Удержание:** Предполагаем 80% для подписок, 40% для Pay-per-Use (разовые пользователи реже возвращаются)

### **Обоснование цифр:**

**Новые пользователи:**

Месяцы 1-3 (PoC): 500 → 700 → 1,300 (рост за счёт сарафанного радио, тестирования в колледже)

Месяцы 4-6 (MVP): 2,000 → 2,500 → 3,000 (запуск таргетированной рекламы с бюджетом 50к/мес)

Месяцы 7-12 (Pilot): 5,000 → 8,000 → 10,000 (масштабирование маркетинга, первые кейсы, SEO-трафик)

**Конверсии:**

97% остаются бесплатными в первый месяц — высокий показатель, так как даём хорошую ценность в бесплатном формате

4% конвертятся в Pay-per-Use — реалистично для низкого барьера входа (100 руб.)

3-4% в подписки Basic/Pro — типично для SaaS с ценностью доказанной через Pay-per-Use

0.01-0.03% в B2B — низкая конверсия из общего трафика, но высокая из целевых обращений

**Удержание (churn rate):**

Pay-per-Use: 60% уходят после первого запроса, 40% возвращаются

Подписки: 80% продлевают (выше среднего по рынку благодаря ценности данных)

B2B: 90% продлевают (высокая лояльность при успешной интеграции)
